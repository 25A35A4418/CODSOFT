```python
print("🤖 Rule-Based Chatbot")
print("Type 'bye' to exit.\n")

while True:

    user = input("You: ").lower()

    if user in ["hi", "hello", "hey"]:
        print("Bot: Hello! How can I help you?")

    elif "name" in user:
        print("Bot: My name is RuleBot.")

    elif "how are you" in user:
        print("Bot: I am fine. Thank you!")

    elif "course" in user:
        print("Bot: I can help with AI and Machine Learning topics.")

    elif "python" in user:
        print("Bot: Python is a popular programming language for AI.")

    elif "thank you" in user or "thanks" in user:
        print("Bot: You're welcome!")

    elif user == "bye":
        print("Bot: Goodbye! Have a nice day.")
        break

    else:
        print("Bot: Sorry, I don't understand that.")
```

```python
import pandas as pd

df = pd.read_csv("chatbot_intent_classification.csv")

qa_dict = dict(zip(df["user_input"], df["intent"]))

print("Dataset Chatbot Started")

while True:

    user = input("You: ")

    if user.lower() == "bye":
        print("Bot: Goodbye!")
        break

    response = qa_dict.get(
        user,
        "Sorry, I don't know the answer."
    )

    print("Bot:", response)
```
